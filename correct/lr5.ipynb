```python
def find_interval(avg_x, x_vals, y_vals):
    avg_x = int(avg_x)
    for i in range(len(x_vals) - 1):
        if (x_vals[i] >= avg_x >= x_vals[i + 1]) or (x_vals[i] <= avg_x <= x_vals[i + 1]):
            return [x_vals[i], x_vals[i + 1]], [y_vals[i], y_vals[i + 1]]
```

```python
points_x = x3
points_y = y

print(points_x)
print(points_y)
```

```python
start_point = [points_x[0], points_y[0]]
end_point = [points_x[-1], points_y[-1]]


models = {'y = ax + b': [],
        'y = ax^b': [],
        'y = ab^x': np.zeros(4),
        'y = a + b/x': np.zeros(4),
        'y = 1/(ax+b)': np.zeros(4),
        'y = x/(ax+b)': np.zeros(4),
        'y = a ln x + b': np.zeros(4)}

for model, pts_container in models.items():
    if model == 'y = ax + b':
        avg_x = (start_point[0] + end_point[0]) / 2
        avg_y = (start_point[1] + end_point[1]) / 2
        x_s = avg_x
    if model == 'y = ax^b':
        avg_x = math.sqrt(start_point[0] * end_point[0])
        avg_y = math.sqrt(start_point[1] * end_point[1])
        x_s = avg_x
    if model == 'y = ab^x':
        avg_x = (start_point[0] + end_point[0]) / 2
        avg_y = math.sqrt(start_point[1] * end_point[1])
        x_s = avg_x
    if model == 'y = a + b/x':
        avg_x = (2 * start_point[0] * end_point[0]) / (start_point[0] + end_point[0])
        avg_y = (start_point[1] + end_point[1]) / 2
        x_s = avg_x
    if model == 'y = 1/(ax+b)':
        avg_x = (start_point[0] + end_point[0]) / 2
        avg_y = (2 * start_point[1] * end_point[1]) / (start_point[1] + end_point[1])
        x_s = avg_x
    if model == 'y = x/(ax+b)':
        avg_x = (2 * start_point[0] * end_point[0]) / (start_point[0] + end_point[0])
        avg_y = (2 * start_point[1] * end_point[1]) / (start_point[1] + end_point[1])
        x_s = avg_x
    if model == 'y = a ln x + b':
        avg_x = math.sqrt(start_point[0] * end_point[0])
        avg_y = (start_point[1] + end_point[1]) / 2
        x_s = avg_x
    
    x_inter, y_inter = find_interval(avg_x, points_x, points_y)
    
    if int(avg_x) in points_x:
        index_i = np.where(points_x == int(avg_x) )[0][0]
        y_s = points_y[index_i]
    else:
        y_s = y_inter[0] + (y_inter[1] - y_inter[0]) / (x_inter[1] - x_inter[0]) * (x_s - x_inter[0])
    
    models[model] = [round(num, 3) for num in [avg_x, avg_y, y_s, abs(avg_y - y_s)]]
```
