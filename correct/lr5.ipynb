```python
c_mat = c.to_numpy()
c_mat_t = c_mat.T 


res=np.dot(c_mat_t , c_mat)
res = np.linalg.inv(res)
res = np.dot(res, c_mat_t)
res = np.dot(res, y)

res
```

```python
def regr_func(x1,x2,x3,x4,x5):
    return res[0] + res[1] * x1 + res[2] * x2 + res[3] * x3 + res[4] * x4 + res[5] * x5
```

```python
y_prognoz_df = pd.DataFrame( columns=[ 'y_prognoz'])
y_prognoz = []

for row_id in range(len(df['x1'])):
    x1_val = df['x1'][row_id]    
    x2_val = df['x2'][row_id]    
    x3_val = df['x3'][row_id]    
    x4_val = df['x4'][row_id]    
    x5_val = df['x5'][row_id]    
    y_res = regr_func(x1_val, x2_val, x3_val, x4_val, x5_val)
    y_prognoz.append(y_res)

y_prognoz_df.y_prognoz = y_prognoz
y_prognoz_df
```

```python
df['y'].plot()
y_prognoz_df['y_prognoz'].plot()
plt.legend(["оригинал", "прогноз"])
```

```python
ostatki_df = pd.DataFrame( columns=[ 'ostatki'])
ostatki = []

for row_id in range(len(df['x1'])):
    y_prognoz = y_prognoz_df['y_prognoz'][row_id]    
    y_real = df['y'][row_id]    

    ostatki.append(y_real - y_prognoz)

ostatki_df.ostatki = ostatki
ostatki_df
```

```python
y_avg = count_avg(df['y'])

otklonenie_df = pd.DataFrame( columns=[ 'otklonenie'])
otklonenie = []

for row_id in range(len(df['x1'])):
    ostatki.append(df['y'][row_id]  - y_avg)

otklonenie_df.otklonenie = ostatki
otklonenie_df
```

```python
# коэф детерминации

ostatki_mat = ostatki_df.to_numpy()
otklonenie_mat = otklonenie_df.to_numpy()


r2 =   1-( np.dot(ostatki_mat.T, ostatki_mat) / np.dot(otklonenie_mat.T, otklonenie_mat) )

r2_adj = 1 - ( np.dot(ostatki_mat.T, ostatki_mat) ** 2 / np.dot(otklonenie_mat.T, otklonenie_mat) ** 2)

print('r2=',r2, 'r2_adj=',r2_adj)
```
