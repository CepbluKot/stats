```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math


data = None
with open('Москва_2021.txt') as file:
    data = [int(line.rstrip()) for line in file]

# print(data)
```

Вычислить групповые средние и дисперсии.

```python
def count_mathematical_expectation(data: list):
    num_of_occurrences = {}

    for val in data:
        if val not in num_of_occurrences:
            num_of_occurrences[val] = 1
        else:
            num_of_occurrences[val] += 1 

    mat_exp = 0

    for val in num_of_occurrences:
        mat_exp += (num_of_occurrences[val] / len(data)) * val

    return mat_exp


def count_dispersion(data: list):
    num_of_occurrences = {}

    for val in data:
        if val not in num_of_occurrences:
            num_of_occurrences[val] = 1
        else:
            num_of_occurrences[val] += 1 

    mat_exp = count_mathematical_expectation(data)
    dispersion = 0

    for val in num_of_occurrences:
        dispersion += (val - mat_exp) ** 2 * (num_of_occurrences[val] / len(data))
    
    return dispersion

def count_avg(data: list):
    return sum(data) / len(data)
```

```python
n_groups = 1 + math.log2(len(data))

print('число групп: ', n_groups)
n_groups= math.ceil(n_groups)
print('после округления: ', n_groups)
```

```python
counts, bins = np.histogram(data, bins=n_groups)
for val_id in range(len(bins)):
    bins[val_id] = int(bins[val_id])

groups_data = [[] for _ in range(n_groups)]
print(bins)
```

```python
def split_data_to_bin(data:list, begin: int, end: int):
    splitted = []
    for val in data:
        if val >= begin and val <= end:
            splitted.append(val)
    
    return splitted

for bin_id in range(len(bins)-1):
    splitted = split_data_to_bin(data, bins[bin_id], bins[bin_id+1])
    groups_data[bin_id] = sorted(splitted)

for gr_id in range(len(groups_data)):
    print(gr_id+1, ". ", "группа ", groups_data[gr_id][0], " - ", groups_data[gr_id][-1], " лет: ", (groups_data[gr_id]))
```

```python
for gr_id in range(len(groups_data)):
    print(gr_id+1, ". ", "группа ", groups_data[gr_id][0], " - ", groups_data[gr_id][-1])
    avg = count_avg(groups_data[gr_id])
    dispersion = count_dispersion(groups_data[gr_id])
    
    print('среднее: ', avg, " дисперсия: ", dispersion)
    print()

```

2. Убедиться, что общая дисперсия равна сумме внутригрупповой и
межгрупповой дисперсии.

```python
full_dispersion = count_dispersion(data)
all_group_dispersions = []

for gr_id in range(len(groups_data)): 
    avg = count_avg(groups_data[gr_id])
    dispersion = count_dispersion(groups_data[gr_id])
    all_group_dispersions.append(dispersion)


print('дисперсия всей выборки: ', full_dispersion)
print('сумма дисперсий групп: ', sum(all_group_dispersions))
```
